# 2. Recursive function
Evaluate the activity of the recursive function? 

```python
def recursive_function(number:int):
  if number <= 1:
    pass
  elif number % 2 == 0:
    return recursive_function(int(number * 2 / 3))
  else:
    return recursive_function(number // 2)

recursive_function(13)
```

# 3. Strange Function
Evaluate the activity of the strange_function? 

```python
def strange_function(number:int, nr_count:int = 1):
  if number <= 1:
    return nr_count
  elif number % 2 == 0:
    return strange_function(int(number * 2 / 3), nr_count + 1)
  else:
    return strange_function(number // 2, nr_count + 1)


strange_function(221412)
```
